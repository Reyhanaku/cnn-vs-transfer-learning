### 1. Analisis Dataset

**1. Apakah dataset yang digunakan cukup besar untuk melatih CNN from scratch?**
Untuk model *CNN from scratch*, dataset CIFAR-10 yang digunakan sudah sangat memadai karena kita mengekstrak ribuan gambar untuk dua kelas (pesawat dan mobil). Jumlah data yang besar ini memberikan ruang yang cukup bagi model untuk mempelajari bobot dari awal. Sebaliknya, pada eksperimen *Transfer Learning*, kita sengaja membatasi dataset menjadi 300 gambar untuk membuktikan bahwa model *pretrained* mampu beroperasi optimal pada skenario data yang sangat minim.

**2. Apakah variasi gambar sudah cukup baik?**
Ya, variasi gambar pada kedua dataset tergolong baik. Pada CIFAR-10, objek difoto dari berbagai sudut (misalnya pesawat tampak depan, samping, atau terbang). Pada dataset Cats vs Dogs, variasi postur hewan, ras hewan, dan jarak pemotretan sangat beragam.

**3. Apakah terdapat ketidakseimbangan data?**
Tidak ada ketidakseimbangan data (*data imbalance*). Pada tahap *preprocessing*, kita telah menggunakan parameter `stratify` saat membagi data latih, validasi, dan uji. Selain itu, pada dataset Cats vs Dogs, kita secara spesifik dan seimbang mengambil tepat 150 gambar kucing dan 150 gambar anjing.

**4. Apakah gambar memiliki noise, background kompleks, atau variasi pencahayaan?**
Gambar pada dataset Cats vs Dogs memiliki latar belakang yang sangat kompleks (hewan di atas kasur, di luar ruangan, atau terhalang objek lain) serta pencahayaan yang bervariasi. Sementara itu, CIFAR-10 memiliki resolusi yang sangat rendah (32x32 piksel), yang secara alami bertindak seperti "noise" yang memaksa model untuk fokus pada pola makro alih-alih detail kecil.

**5. Bagaimana kualitas dataset memengaruhi hasil CNN dan transfer learning?**
Kualitas dan kompleksitas dataset sangat memengaruhi performa. Resolusi rendah pada CIFAR-10 membuat *CNN from scratch* harus berjuang mengekstrak fitur, sehingga akurasinya butuh waktu untuk naik. Sebaliknya, gambar beresolusi tinggi (160x160) dengan latar belakang kompleks pada Cats vs Dogs bisa ditangani dengan sangat mudah oleh *Transfer Learning* karena model tersebut sudah pernah dilatih dengan jutaan gambar kompleks dari ImageNet.

### 2. Analisis Performa Model

**1. Model mana yang menghasilkan performa terbaik?**
Secara keseluruhan, model *Transfer Learning* (MobileNetV2) menghasilkan performa terbaik. Akurasi pelatihannya jauh lebih tinggi, konvergensinya tercapai lebih cepat (hanya dalam beberapa *epoch* awal), dan *loss*-nya lebih rendah dibandingkan *CNN from scratch*.

**2. Apakah akurasi tinggi selalu berarti model lebih baik?**
Tidak selalu. Akurasi tinggi yang dibarengi dengan ukuran model yang terlalu raksasa (jutaan parameter) atau waktu inferensi yang lambat bisa menjadi tidak berguna jika diimplementasikan pada perangkat dengan sumber daya terbatas (seperti *mobile* atau *IoT*). Model yang baik adalah model yang seimbang antara akurasi, kecepatan, dan efisiensi memori.

**3. Apakah ada tanda-tanda overfitting?**
*(Catatan: Sesuaikan kalimat ini dengan grafikmu. Pilih salah satu)*
*Opsi 1:* Ya, terdapat indikasi *overfitting* ringan pada *CNN from scratch*, terlihat dari akurasi *training* yang terus meningkat mendekati 100%, namun akurasi *validation* cenderung stagnan atau *loss validation*-nya mulai naik.
*Opsi 2:* Tidak ada tanda *overfitting* yang parah berkat penggunaan teknik regularisasi `Dropout` pada arsitektur model, sehingga garis validasi tetap mengikuti pergerakan garis *training*.

**4. Model mana yang lebih stabil selama training?**
*Transfer Learning* jauh lebih stabil. Grafik akurasi dan *loss*-nya bergerak mulus tanpa fluktuasi tajam. Hal ini terjadi karena lapisan ekstraksi fiturnya sudah di-*freeze* (dibekukan), sehingga *optimizer* hanya perlu menstabilkan satu lapisan *Dense* di akhir.

**5. Bagaimana hubungan antara jumlah data dan performa model?**
Hubungannya sangat berbanding lurus, terutama untuk *CNN from scratch*. Semakin banyak data dengan variasi tinggi, model akan semakin tangguh menghadapi pola baru. Namun, *Transfer Learning* mematahkan aturan ini dengan kemampuannya memberikan performa maksimal meskipun hanya disuntikkan sedikit data baru.
